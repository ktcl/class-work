## Simple Chart Plotting Examples

```python
# Start pandas and matplotlib - don't forget the % or inline
import pandas as pd
%matplotlib inline
```

```python
df = pd.DataFrame({'Sales':[65,47,81,55,73]})
df
```

```python
# Simple default plot
df.plot()
```

```python
# adding a ; gets rid of <matplotlib.axes._subplots.AxesSubplot at 0x1786e1ab6d8>
df.plot();
```

```python
# also gets rid of <matplotlib.axes._subplots.AxesSubplot at 0x1786e1ab6d8>
p = df.plot()
```

```python
# creates a bar chart
p = df.plot(kind = 'bar')
```

```python
# creates a horizontal bar chart
p = df.plot(kind = 'barh')
```

```python
# also creates a horizontal bar chart same for line() and .bar()
p = df.plot.barh()
```

```python
# not all shart kinds will work with our data
p = df.plot.area(figsize=(15,5))
```

```python
# to add chart features. Pressing the [Tab] key after typing p.set brings up list of options.
p = df.plot.line(figsize=(15,5))
p.set_title('Department Sales')
```

```python
# To change the title font size. remember ; gets rid of text in output Text(0.5,1,'Department Sales')
p = df.plot.line(figsize=(15,5))
p.set_title('Department Sales', fontsize=40);
p.set_xlabel('X Axis',fontsize=15);
p.set_ylabel('Y Axis',fontsize=15);
```

```python
# To change the legend
p = df.plot.line(figsize=(15,5))
p.set_title('Department Sales', fontsize=40);
p.set_xlabel('X Axis',fontsize=15);
p.set_ylabel('Y Axis',fontsize=15);
p.legend(['New Legend']);
```

```python
# To change the color of the plot line and the title. Colors are not case sensitive.
p = df.plot.line(figsize=(15,5), color='red')
p.set_title('Department Sales', fontsize=40, color='Purple');
p.set_xlabel('X Axis',fontsize=15);
p.set_ylabel('Y Axis',fontsize=15);
p.legend(['New Legend']);
```
